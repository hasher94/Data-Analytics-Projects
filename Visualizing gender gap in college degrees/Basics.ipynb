```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering', 'Computer Science']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
fig = plt.figure(figsize=(16,20))

for sp in range(0,18,3):
    indx = int(sp/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[indx]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[indx]], c=cb_orange, label='Men', linewidth=3)
    for key,value in ax.spines.items():
        value.set_visible(False)    
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[indx])
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom ='off')
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255,171/255,171/255), alpha=0.3)
    
    if indx == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif indx == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        ax.tick_params(labelbottom='on')
        
for sp in range(1,19,3):
    indx = int((sp-1)/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[indx]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[indx]], c=cb_orange, label='Men', linewidth=3)
    for key,value in ax.spines.items():
        value.set_visible(False)    
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(other_cats[indx])
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom='off')
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255,171/255,171/255), alpha=0.3)
    
    if indx == 0:
        ax.text(2005, 92, 'Women')
        ax.text(2002, 8, 'Men')
    elif indx == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        ax.tick_params(labelbottom='on')
        
for sp in range(2,17,3):
    indx = int((sp-2)/3)
    ax = fig.add_subplot(6,3,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[indx]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[indx]], c=cb_orange, label='Men', linewidth=3)
    for key,value in ax.spines.items():
        value.set_visible(False)    
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(lib_arts_cats[indx])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    ax.set_yticks([0,100])
    ax.axhline(50, c=(171/255,171/255,171/255), alpha=0.3)
    
    if indx == 0:
        ax.text(2005, 75, 'Women')
        ax.text(2001, 35, 'Men')
    if indx == 4:
        ax.tick_params(labelbottom='on')
plt.show()
```
